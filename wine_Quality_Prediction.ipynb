```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

df=pd.read_csv(r"C:\Users\HP\Desktop\Oasis internship\Wine quality prediction project2lvl2\WineQT.csv")
df.head()
```

```python
df.info()
df.shape
df.describe()
```

```python
df.isnull().sum()
```

```python
features = ['density', 'fixed acidity', 'volatile acidity', 'citric acid']
X = df[features]
y = df['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=11
)
```

```python
models = {
    "Random Forest": RandomForestClassifier(random_state=11),
    "SGD Classifier": SGDClassifier(random_state=11, max_iter=1000, tol=1e-3),
    "SVC": SVC(random_state=42)
}
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"\n{name} Results:")
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
```
